## Calculating Correlation Matrix

```python
#Dropping the target variable

data = data.drop(['Purchase'], axis = 1)
```

```python
#Dropping the Categorical Variable
data = data.drop(['Product_ID'], axis = 1)
```

```python
#Encoding the Age variable using map function
data['Age'] = data['Age'].map({'0-17': 1 , '18-25': 2 , '26-35': 3 , '36-45': 4 , '46-50': 5 , '51-55': 6 , '55+': 6})
```

```python
#Encoding the city category variable

data['City_Category'] = data['City_Category'].map({'A': 1 , 'B': 2 , 'C': 3})
```

```python
#Encoding the Stay in city variable as it is an object datatype

data['Stay_In_Current_City_Years'] = data['Stay_In_Current_City_Years'].map({'0': 0 , '1': 1 , '2': 2 , '3': 3 , '4+': 4})
```

```python
#Encoding the Gender Variable

data = pd.get_dummies(data, drop_first = True)
```

```python
data.head()
```

```python
# Correlation matrix for independent variables
data.corr()
```

```python
#Finding the absolute values and only the upper diagonal of the correlation matrix

matrix = data.corr().abs()
upper_diagonal = matrix.where(np.triu(np.ones(matrix.shape), k=1).astype(np.bool))
upper_diagonal
```

#### Looking at the above values, we can say that Product_Category_2, Product_Category_3 are highly correlated with correlation value 0.416680 & Product_Category_1, Product_Category_2 are highly correlated with correlation value 0.331691.

## Covariance Matrix 

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
data_scale = scale.fit_transform(data)
```

```python
data_scale = pd.DataFrame(data_scale, columns = data.columns)
```

```python
data_scale.head()
```

```python
data_scale.cov()
```

```python
# Checking the upper diagonal of the covariance matrix

covar_matrix = data_scale.cov()
upper_dia = covar_matrix.where(np.triu(np.ones(covar_matrix.shape), k=1).astype(np.bool))
upper_dia
```

## Eigen Values and Eigen Vectors

```python
# Calculating Eigen values and Eigen vectors from the Covariance Matrix
from numpy.linalg import eig

eigen_value, eigen_vector = eig(covar_matrix)
```

```python
print('Eigen values are - ', eigen_value)
print('\nEigen vectors are - ', eigen_vector)
```

### Selecting Principal Components

```python
#Arranging the magnitude of eigen values in decreasing order

sort_index = np.argsort(eigen_value)[::-1]
sort_eigen_value = eigen_value[sort_index]
sort_eigen_value
```

```python
#Arranging the eigen vectors

sort_eigen_vector = eigen_vector[:, sort_index]
sort_eigen_vector
```

```python
#Calculating the percentage of variance of each eigen vector

for j in sort_eigen_value:
    print(j / sum(sort_eigen_value))
```

```python
#Selecting the most 2 important Principal Components
k_components = 2
eigen_vec_subset = sort_eigen_vector[:,0:k_components]
eigen_vec_subset
```

### Selecting 8 features using Principal Component Analysis

```python
#Selecting 8 principal components

k_components = 8
eigen_vec_subset = sort_eigen_vector[:,0:k_components]
eigen_vec_subset
```
